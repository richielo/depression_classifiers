```python
from __future__ import print_function
from IPython.display import display, HTML
import os
import sys
import csv
import time
import statistics
import numpy as np
import string
import re
import pandas as pd
import text_features_extractor as tfExtractor
```

```python
#Utility functions
def mergeTwoDicts(x, y):
    """Given two dicts, merge them into a new dict as a shallow copy."""
    z = x.copy()
    z.update(y)
    return z
```

```python
#Read in cleaned data csv
chinCleanedDf = pd.read_csv('cleaned_data/chin_data_cleaned.csv', encoding='utf-8')
engCleanedDf = pd.read_csv('cleaned_data/eng_data_cleaned.csv', encoding='utf-8')
```

#### Extraction for Chinese

```python
#Convert text column to list and sentiment column to list
chinTextList = chinCleanedDf['content'].tolist()
chinSentLabels = chinCleanedDf['depressed'].tolist()
```

```python
#Extract Liwc
chinLiwcResults = tfExtractor.batchLiwcRun(chinTextList, 'raw', filters=[])
print("Liwc result length: " + str(len(chinLiwcResults)))
```

```python
#Extract Anew
chinAnewResults = tfExtractor.batchAnewRun(chinTextList, 'raw')
print("Anew result length: " + str(len(chinAnewResults)))
```

```python
#Extract punctuation stats
chinPuncResults = []
for text in chinTextList:
    chinPuncResults.append(tfExtractor.getPuncStats(text))
print("Punc stats result length: " + str(len(chinPuncResults)))
```

```python
#Combine the results
mergedResult = []
i = 0
for r in chinLiwcResults:
    anewDict = chinAnewResults[i]
    del anewDict['triggers']
    mergedDict = mergeTwoDicts(mergeTwoDicts(r['stats'], anewDict), chinPuncResults[i])
    mergedDict['depressed'] = chinSentLabels[i]
    mergedResult.append(mergedDict)
    i += 1
#Convert to dataframe then csv
chinFinalDf = pd.DataFrame(mergedResult)
chinFinalDf.to_csv('data_with_features/chin_cleaned_data_f.csv', encoding='utf-8')
```

#### Extraction for English

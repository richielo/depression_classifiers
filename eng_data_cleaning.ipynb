```python
from __future__ import print_function
from IPython.display import display, HTML
import os
import sys
import csv
import time
import statistics
import numpy as np
import string
import pandas as pd
import json
```

```python
#import and parse data, match with labels
rawEngTwitData = open('data/tweetdata.txt', 'r')
twitDataLabels = pd.read_excel('data/tweetlabels.xlsx', sheetname='sentiment2')
twitDataListDict = []
i = 0
for line in rawEngTwitData:
    try:
        twitObj = json.loads(line)
    except ValueError, e:
        continue
    #Empty object
    if('created_at'not in twitObj):
        continue
    else:
        twitDataRowDict = {}
        #Get data fields
        i += 1
        twitID = twitObj['id']
        twitIDString = twitObj['id_str']
        twitText = twitObj['text']
        #print('twitID: ' + str(twitID))
        #Find corresponding label from label dataframe
        labelRows = twitDataLabels[twitDataLabels.id == twitID]
        if(len(labelRows.index) > 0):
            label = labelRows['sentiment'].iloc[0]
            if(label == 0):
                label = -1
            elif(label == 1):
                label = -1
            elif(label == -1):
                label = 1
            #Store extracted data
            twitDataRowDict['ID'] = i
            #twitDataRowDict['TwitIDString'] = twitIDString
            twitDataRowDict['content'] = twitText
            twitDataRowDict['depressed'] = label
            twitDataListDict.append(twitDataRowDict)
print("Number of rows: " + str(len(twitDataListDict)))
```

```python
#Convert to dataframe and save as csv
twitDataDf = pd.DataFrame(twitDataListDict)
twitDataDf.to_csv('cleaned_data/eng_data_cleaned.csv', encoding='utf-8')
display(twitDataDf)
```

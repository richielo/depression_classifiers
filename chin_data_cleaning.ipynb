```python
import os
import sys
import numpy as np
import string
import pandas as pd
from IPython.display import display, HTML
from hanziconv import HanziConv
from zhon import hanzi
import re
import jieba
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
```

```python
def removePuncFromContext(contextList):
    nopuncContextList = []
    for context in contextList:
        if(context not in stop):
            nopuncContextList.append(context)
    return nopuncContextList

def removeChinStopwords(context):
    # read stopwords list from local file
    stop_f = open('data/stopwords-zh.txt','r')
    stopwords = [l.strip() for l in stop_f.readlines()]
    for i in range(len(stopwords)):
        stopwords[i] = unicode(stopwords[i], "utf8", errors="ignore")
    #print(stopwords)
    stop_f.close()
    clean = [t for t in context if t not in stopwords]
    return clean

def chinPreprocessing(text):
    simChinText = HanziConv.toSimplified(text.replace(' ', ''))
    segList = jieba.cut(simChinText, cut_all=False)
    cleanedText = removePuncFromContext(removeChinStopwords(segList))
    return cleanedText
```

```python
def splitChinEng(text):
    text = unicode(text, 'utf-8')
    foundChin = re.findall('[{}]'.format(hanzi.characters.encode('utf-8')), text.encode('utf-8'))
    foundEng =  re.sub("[^A-Za-z]", " ", text.encode('utf-8').strip())
    
    print("Found Chin:")
    print(" ".join(chinPreprocessing(''.join(foundChin))))
    print("Found Eng:")
    print(" ".join(foundEng.split()))
```

```python
#Import raw data to be cleaned and parsed
#For positive data
rawBlogPosData = open('data/raw_yes_parsed_then_merged_17-2-2014.txt')
rawBlogPosListDict = []
currRawBlogPosDict = {}
currRawBlogText = ""
currID = -1
skipTitle = False 
for line in rawBlogPosData:
    if(line == '\n'):
        continue
    noWhiteSpace = line.replace(' ', '')
    if('id=' in noWhiteSpace and 'http' not in noWhiteSpace):
        #Store present entry
        if(currID != -1 ):
            currRawBlogPosDict['content'] = currRawBlogText
            currRawBlogPosDict['depressed'] = 1
            rawBlogPosListDict.append(currRawBlogPosDict)
        #Update for new entry
        currID = int(re.search(r'\d+', noWhiteSpace).group())
        print("CurrID: " + str(currID))
        currRawBlogPosDict = {}
        currRawBlogPosDict['ID'] = currID
        currRawBlogText = "" 
    else:
        currRawBlogText += line 
#Add last blog
currRawBlogPosDict['ID'] = currID
currRawBlogPosDict['content'] = currRawBlogText
currRawBlogPosDict['depressed'] = 1
rawBlogPosListDict.append(currRawBlogPosDict)
```

```python
#Import raw data to be cleaned and parsed
#For positive data
rawBlogNegData = open('data/raw_no_parsed_then_merged_17-2-2014.txt')
rawBlogNegListDict = []
currRawBlogNegDict = {}
currRawBlogText = ""
currID = -1
skipTitle = False 
for line in rawBlogNegData:
    if(line == '\n'):
        continue
    noWhiteSpace = line.replace(' ', '')
    if('id=' in noWhiteSpace and 'http' not in noWhiteSpace):
        #Store present entry
        if(currID != -1 ):
            currRawBlogNegDict['content'] = currRawBlogText
            currRawBlogNegDict['depressed'] = -1
            rawBlogNegListDict.append(currRawBlogNegDict)
        #Update for new entry
        currID = int(re.search(r'\d+', noWhiteSpace).group())
        print("CurrID: " + str(currID))
        currRawBlogNegDict = {}
        currRawBlogNegDict['ID'] = currID + 274
        currRawBlogText = "" 
    else:
        currRawBlogText += line 
#Add last blog
currRawBlogNegDict['ID'] = currID + 274
currRawBlogNegDict['content'] = currRawBlogText
currRawBlogNegDict['depressed'] = -1
rawBlogNegListDict.append(currRawBlogNegDict)
```

```python
#Merge the to list-dicts to 1, convert to dataframe and output as csv
rawBlogListDict = rawBlogPosListDict + rawBlogNegListDict
rawBlogDf = pd.DataFrame(rawBlogListDict)
display(rawBlogDf)
rawBlogDf.to_csv('cleaned_data/chin_data_cleaned.csv', encoding='utf-8')
```
